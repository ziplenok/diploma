```python
import pandas as pd #для работы с датасетами
import numpy as np #для работы с числами
from sklearn import preprocessing #для исполь. label encoder, который преобразует текст в числа, чтоб моделька легче училась и могла принимать эти данные, в идеале было бы лучше использовать onehot encoder, но потом я не разберусь что где, мб запутаюсь
```

```python
df = pd.read_csv('df_final_telegram.csv') # считываем файл и сохраняем его в переменную df
df['profile_subjects'] = df['profile subject_1']+df['profile subject_2'] #соеденнил 2 колонны с профильными предметами, дабы проблем никаких не возникло при генерации и обучении
```

##Создание фэйковых данных на основе рандома и логики


```python
df.drop(columns=['Unnamed: 0', 'index','Univer_code', 'profile subject_1', 'profile subject_2', '2020-2021_Max', '2020-2021_Min', '2021-2022_Max', '2021-2022_Min', 'difference_Max', 'difference_Min', 'Count_difference_per_year'], inplace=True)
# df_decoded = df.copy()
#убрал столбцы, которые я посчитал ненужными, возможно верну некоторые в будущем
```

```python
le = preprocessing.LabelEncoder() #тот самый label encoder
df['specialization_name_encoded']=le.fit_transform(df['specialization_name']) #создается новый столбец уже преобразованный в числа от основного столбца
df.drop(['specialization_name'], axis=1, inplace=True) #дропнул основной текстовой столбец
df['Status_encoded']=le.fit_transform(df['Status'])             # дальше то же самое с несколькими столбцами, так как они текстовые или категоричные
df.drop(['Status'], axis=1, inplace=True)                       # позже уберу лэйбл енкодер с некоторых столбцов, таких как суммабалловаттестата и средний бал аттестата
df['преимущественноеправо_encoded']=le.fit_transform(df['преимущественноеправо'])
df.drop(['преимущественноеправо'], axis=1, inplace=True)
df['среднийбаллаттестата(диплома)_encoded']=le.fit_transform(df['среднийбаллаттестата(диплома)'])
df.drop(['среднийбаллаттестата(диплома)'], axis=1, inplace=True)
df['суммабалловпрофильныхпредметов(творческихэкзаменов)_encoded']=le.fit_transform(df['суммабалловпрофильныхпредметов(творческихэкзаменов)'])
df.drop(['суммабалловпрофильныхпредметов(творческихэкзаменов)'], axis=1, inplace=True)
df['profile_subjects_encoded']=le.fit_transform(df['profile_subjects'])
df.drop(['profile_subjects'], axis=1, inplace=True)
```

```python
def minMax(x):  #для нахождения минимума и максимума из столбца
    return pd.Series(index=['min','max'],data=[x.min(),x.max()])
```

#Для создания логики надо сделать анализ существующих данных


```python
# группировка датасета по столбцу выборки и применение методов min и max
min_max_specs = df.groupby('specialization_name_encoded').agg({'Points': ['min', 'max']})

# принт, просто принт
print(min_max_specs)

#это нужно для того, чтоб примерно знать пороговый балл грантников, по идее в таком слуачае если будет огромное количество данных, то и точный порог сможем найти
```

```python
df_new['grant']=0 # новый столбец с нулями
```

```python
for i in range(df['Points'].size): #этот цикл перебирает по количеству строк в датасете, где i это индекс 
    if not df_new.iloc[i]['Points']<min_max_specs.iloc[df_new.iloc[i]['specialization_name_encoded']]['Points']['min']:  #в этом условии идет сравнение баллов абитуриуента и минимум балла от специальности данной специальности
      df_new.at[i, 'grant'] = 1 
    #если понтс не меньше чем минимум, то грант
```
